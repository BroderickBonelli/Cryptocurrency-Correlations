```python
import numpy as np
import pandas as pd
import seaborn as sn
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

btc = pd.read_csv("/Users/broderickbonelli/Desktop/Crypto_correlations/BTC_USD.csv")
btc_df = pd.DataFrame(btc)
btc_df = btc_df.drop(['Currency', '24h Open (USD)', '24h High (USD)', '24h Low (USD)'], axis=1)
btc_df.rename(columns={'Closing Price (USD)':'BTC_USD Closing Price'}, inplace=True)
#print(btc_df)

eth = pd.read_csv("/Users/broderickbonelli/Desktop/Crypto_correlations/ETH_USD.csv")
eth_df = pd.DataFrame(eth)
eth_df = eth_df.drop(['Currency', '24h Open (USD)', '24h High (USD)', '24h Low (USD)'], axis=1)
eth_df.rename(columns={'Closing Price (USD)':'ETH_USD Closing Price'}, inplace=True)
#print(eth_df)

xrp = pd.read_csv("/Users/broderickbonelli/Desktop/Crypto_correlations/XRP_USD.csv")
xrp_df = pd.DataFrame(xrp)
xrp_df = xrp_df.drop(['Currency', '24h Open (USD)', '24h High (USD)', '24h Low (USD)'], axis=1)
xrp_df.rename(columns={'Closing Price (USD)':'XRP_USD Closing Price'}, inplace=True)
#print(xrp_df)

sol = pd.read_csv("/Users/broderickbonelli/Desktop/Crypto_correlations/SOL_USD.csv")
sol_df = pd.DataFrame(sol)
sol_df = sol_df.drop(['Currency', '24h Open (USD)', '24h High (USD)', '24h Low (USD)'], axis=1)
sol_df.rename(columns={'Closing Price (USD)':'SOL_USD Closing Price'}, inplace=True)
#print(sol_df)

luna = pd.read_csv("/Users/broderickbonelli/Desktop/Crypto_correlations/LUNA_USD.csv")
luna_df = pd.DataFrame(luna)
luna_df = luna_df.drop(['Currency', '24h Open (USD)', '24h High (USD)', '24h Low (USD)'], axis=1)
luna_df.rename(columns={'Closing Price (USD)':'LUNA_USD Closing Price'}, inplace=True)
#print(luna_df)

ada = pd.read_csv("/Users/broderickbonelli/Desktop/Crypto_correlations/ADA_USD.csv")
ada_df = pd.DataFrame(ada)
ada_df = ada_df.drop(['Currency', '24h Open (USD)', '24h High (USD)', '24h Low (USD)'], axis=1)
ada_df.rename(columns={'Closing Price (USD)':'ADA_USD Closing Price'}, inplace=True)
#print(ada_df)

dot = pd.read_csv("/Users/broderickbonelli/Desktop/Crypto_correlations/DOT_USD.csv")
dot_df = pd.DataFrame(dot)
dot_df = dot_df.drop(['Currency', '24h Open (USD)', '24h High (USD)', '24h Low (USD)'], axis=1)
dot_df.rename(columns={'Closing Price (USD)':'DOT_USD Closing Price'}, inplace=True)
#print(dot_df)

atom = pd.read_csv("/Users/broderickbonelli/Desktop/Crypto_correlations/ATOM_USD.csv")
atom_df = pd.DataFrame(atom)
atom_df = atom_df.drop(['Currency', '24h Open (USD)', '24h High (USD)', '24h Low (USD)'], axis=1)
atom_df.rename(columns={'Closing Price (USD)':'ATOM_USD Closing Price'}, inplace=True)
#print(atom_df)

df = btc_df.merge(eth_df, on='Date')
df = df.merge(xrp_df, on='Date')
df = df.merge(sol_df, on='Date')
df = df.merge(luna_df, on='Date')
df = df.merge(ada_df, on='Date')
df = df.merge(dot_df, on='Date')
df = df.merge(atom_df, on='Date')
#print(df)

df = df.tail(30)
correlation = df.corr()


figure(figsize=(7, 6))
sn.heatmap(correlation, cmap='winter')
plt.title('Crypto Correlation')
plt.show()

display(correlation)
```
